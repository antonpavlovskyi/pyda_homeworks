## Задание 2. Работа с выбросами

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# Найдем межквартльный размах для всех колонок с не категорийными данными, так как выбросы категорийных данных это явные ошибки и для них не корректно расчитывать этот параметр
columns_non_categorical = ['rectal_temperature', 'pulse', 'respiratory_rate', 'nasogastric_reflux_ph', 'packed_cell_volume', 'total_protein', 'abdomcentesis_total_protein']
q1 = horse_data[columns_non_categorical].quantile(0.25)
q3 = horse_data[columns_non_categorical].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
```

```python
# Напишем функцию, которая определит выбросы и отобразит их вместе с колонками 'surgical_lesion', 'type_of_lesion1', 'pain', 'abdominal_distension', которые могут приводить к аномальным значениям
def outliers(column):
    outliers = horse_data[(horse_data[column] < lower_bound[column]) | (horse_data[column] > upper_bound[column])]
    if outliers.empty:
        return 'Эта колонка на имеет выбросов'
    return outliers[[column, 'surgical_lesion', 'type_of_lesion1', 'pain', 'abdominal_distension']]
```

## Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# колонка surgery имеет один пропуск, так как в этой строке значение surgical_lesion = 1 (значит операция была), то можно поставить 1 в surgery
horse_data['surgery'].fillna(value=1, inplace=True)
'Количесто пропусков в колонке surgery: ' + str(len(horse_data[horse_data['surgery'].isna()]))
```

```python
# Создадим колонку type_of_lesion_1_2 и заполним значениями 1 (есть проблема), если type_of_lesion1 или type_of_lesion2 или type_of_lesion3 > 0 и 2 (нет проблемы) , если type_of_lesion1 или type_of_lesion2 или type_of_lesion3 == 0
def type_of_lesion_1_2(row):
    if row['type_of_lesion1'] > 0 or row['type_of_lesion2'] > 0 or row['type_of_lesion3'] >0:
        return 1
    return 2
horse_data['type_of_lesion_1_2'] = horse_data.apply(type_of_lesion_1_2, axis=1)
```

```python
# Функция для заполнения пропусков медианой
def median_fill(column_to_fill, **groupby_kwargs):
    horse_data[column_to_fill].fillna(horse_data.groupby(list(groupby_kwargs.values()))[column_to_fill].transform('median'), inplace=True)
    missign_values = len(horse_data[horse_data[column_to_fill].isna()])
    return f'Количесто пропусков в колонке {column_to_fill}: {missign_values}'
```

```python
# Функция для заполнения пропусков модой
def mode_fill(column_to_fill, **groupby_kwargs):
    horse_data[column_to_fill] = horse_data.groupby(list(groupby_kwargs.values())).transform(lambda x: x.fillna(x.mode()))
    missign_values = len(horse_data[horse_data[column_to_fill].isna()])
    return f'Количесто пропусков в колонке {column_to_fill}: {missign_values}'
```

```python
# Согласно horese_data.names, rectal_temperature зависит от surgery, temperature_of_extremities, type_of_lesion_1_2. Заполним пропуски медианой (выбросы присутствуют и признаны не выходящими за рамки возможного) в разрезе вышеперечисленных групп
median_fill('rectal_temperature', col1='surgery', col2='temperature_of_extremities', col3='type_of_lesion_1_2')
```
