## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

```python
movies_ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies_ratings.head()
```

```python
# Функция для определения класса фильма на основании рейтинга
def rating_class(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    elif rating >= 4.5:
        return 'высокий рейтинг'
```

```python
# Добвляем столбец class, который содержит класс фильма, основываясь на рейтинге
movies_ratings['class'] = movies_ratings['rating'].apply(rating_class)
movies_ratings.head()
```

## Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
# Данные по регионам России
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
```

```python
keywords = pd.read_csv('ml-latest-small/keywords.csv')
keywords.head()
```

```python
# Функция для определения региона по ключевым словам
def geo_class(keyword):
    for key in geo_data:
        if any(city in keyword for city in geo_data[key]):
            return key
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo_class)
keywords.head()
```

## Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

```python
years = range(1950, 2011)
```

```python
movies_25m = pd.read_csv('ml-25m/movies.csv')
movies_25m.head()
```

```python
ratings_25m = pd.read_csv('ml-25m/ratings.csv')
ratings_25m.head()
```

```python
# Фунуия для определения года выпуска фильма по его названию
def production_year(title):
    for year in years:
        if str(year) in title:
            return int(year)
    return 1900
```

```python
# Добавляем колонку year с годом выпуска фильма
movies_25m['year'] = movies_25m['title'].apply(production_year)
movies_25m.head()
```

```python
# Соединяем датафеймы movies_25m и ratings_25m по колонке movieId
combined = pd.merge(movies_25m, ratings_25m, on='movieId')
combined.head()
```

```python
# Группируем датафейм по году выпуска (колонка year) и сортируем по рейтингу по убыванию
year_descending = combined.groupby('year').mean().sort_values('rating', ascending=False)['rating']
year_descending.head(), year_descending.tail()
```
