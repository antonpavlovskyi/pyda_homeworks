# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('files_for_homework/power.csv')
power.head()
```

```python
# фильтр стран
find_counties = ((power['country'].str.contains('lat', case=False)) | (power['country'].str.contains('lit', case=False)) | (power['country'].str.contains('est', case=False)))

# фильтр категорий
power_categories = ((power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21))

# отфильтрованый датафр
filter_power = power[(2005 < power['year']) & (power['year'] < 2010) & power_categories & (power['quantity'] > 0) & find_counties]
filter_power.head()
```

```python
# сумма потребления по отфильтрованому датафрейму
filter_power['quantity'].sum()
```
